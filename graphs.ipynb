# Graphs Relating Features and House Values

## Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the Dataset

```python
dataset = pd.read_csv('datasets_88705_204267_Real estate.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X[:, 1])
```

## Transaction Date vs House Price of unit area


```python
plt.scatter(X[:80, 0], y[:80], color = 'red')
plt.title('Transaction Date vs House Price of unit area')
plt.xlabel('Transaction Date')
plt.ylabel('House Price of unit area')
plt.show()
```

## House Age vs House Price of unit area


```python
plt.scatter(X[:80, 1], y[:80], color = 'red')
plt.title('House Age vs House Price of unit area')
plt.xlabel('House Age')
plt.ylabel('House Price of unit area')
plt.show()
```

## Distance to nearest MT station vs House Price of unit area

```python
plt.scatter(X[:80, 2], y[:80], color = 'red')
plt.title('Distance to nearest MT station vs House Price of unit area')
plt.xlabel('Distance to nearest MT station')
plt.ylabel('House Price of unit area')
plt.show()
```

## Number of convenience stores vs House Price of unit area

```python
plt.scatter(X[:80, 3], y[:80], color = 'red')
plt.title('Number of convenience stores vs House Price of unit area')
plt.xlabel('Number of convenience stores')
plt.ylabel('House Price of unit area')
plt.show()
```
